```python
import pandas as pd

df = pd.read_csv("contentious subjects classification.csv", sep=';', engine='python')
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import networkx as nx

# 2. Nettoyage et sélection des colonnes
columns_to_keep = ['Broad subject', 'Category / Contentious sector', '(Sub-Category)']
df_clean = df[columns_to_keep].dropna().copy()

for col in df_clean.columns:
    df_clean[col] = df_clean[col].astype(str).str.strip()

# On donne un poids égal à chaque sous-catégorie pour la taille des secteurs
df_clean['count'] = 1 

# 3. Définition d'une palette de couleurs explicite et logique
# Permet d'identifier instantanément le bloc hybride (Climate & Biodiversity)
color_map = {
    'Climate': '#3498db',                  # Bleu
    'Biodiversity': '#2ecc71',             # Vert
    'Climate & Biodiversity': '#1abc9c',    # Turquoise / Cyan (Mélange visuel des deux)
    '?' : '#b2bec3'                        # Couleur par défaut si besoin
}

# 4. Création du graphique Sunburst (sans aucune duplication)
fig = px.sunburst(
    df_clean,
    path=['Broad subject', 'Category / Contentious sector', '(Sub-Category)'],
    values='count',
    color='Broad subject',
    color_discrete_map=color_map,
    title="Visualisation Sunburst Unique (Hiérarchie stricte des sujets)"
)

# 5. Optimisation du rendu visuel et des informations au survol
fig.update_traces(
    textinfo="label",
    hovertemplate="<b>%{label}</b><br>Nombre de sous-catégories : %{value}"
)

fig.update_layout(
    title=dict(
        text="Arborescence Sunburst : Classification des Sujets Contentieux",
        font=dict(size=16) # Syntaxe conforme aux dernières versions de Plotly
    ),
    margin=dict(t=60, l=25, r=25, b=25),
    font=dict(size=11),
    width=900,
    height=900
)

# 6. Affichage
fig.show()
```
