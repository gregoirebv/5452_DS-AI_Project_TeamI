### BERT for topic discovery, exploration, visualization
With 2 clean datasets

```python
# HuggingFace token for BERTopic

import os

hf_token = os.getenv("HUGGING_FACE_TOKEN")

if hf_token is None:
    raise ValueError("HUGGING_FACE_TOKEN is not set in the environment")
```

```python
# articles dataset import

import pandas as pd

df2 = pd.read_csv("clean-snopes_checked_v02.csv")
```

```python
# columns check

df2 = df2.dropna(subset=["ArticleText"])

df2 = df2[df2["ArticleText"].str.strip() != ""] # data cleaning: removing rows with empty ArticleText

print(df2.columns)
```

```python
# articles loading check

docs2 = df2["ArticleText"].astype(str).tolist()

print(f"Dataset 2: {len(docs2)} articles")
```

```python
# removing English stop words (the, of, and, to, etc.) using sklearn library (https://github.com/SciSharp/scikit-learn.net/tree/master?tab=readme-ov-file) and following advice from: https://maartengr.github.io/BERTopic/faq.html#how-do-i-reduce-topic-outliers

from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(
    stop_words="english",
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.95
)
```

```python
#training BERTopic and fitting it to Dataset 2 (Snopes)

from bertopic import BERTopic

topic_model = BERTopic(
    vectorizer_model=vectorizer_model,
    language="english",
    min_topic_size=5,
    calculate_probabilities=False,
    verbose=True
)

topics, probs = topic_model.fit_transform(docs2)
```

```python
# Discovered topics

topic_info = topic_model.get_topic_info()
print(topic_info.head(20))
```

```python
# Non-outlier topics

for topic_id in topic_info["Topic"][:5]:
    
    if topic_id == -1:
        continue

    print(f"\nTOPIC {topic_id}")
    print(topic_model.get_topic(topic_id))
```

```python
# Visualization of topics

fig = topic_model.visualize_topics()
fig.show()
```
