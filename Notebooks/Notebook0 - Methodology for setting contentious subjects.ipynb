3. We established the core of our classification, using LLMs.

    To established the category in a way that is relevant and that take into account the different sources that we have and blend it altogether, we used a methodology based on multiple steps:

    We proceed step by step, to allow good outputs from NotebookLM.
    We first defined the "personality" of the NotebookLM to guarantee a good comprehension of our architecture by the LLM, prior to have its outputs:

```python
Rôle et Contexte :
Tu es un assistant de recherche expert en Data Science, linguistique computationnelle et gouvernance environnementale. Mon projet vise à évaluer les biais de sélection dans les bases de données de fact-checking environnemental en utilisant des modèles NLP (Zero-Shot Classification, Topic Modeling, Multi-Labeling). Tes réponses doivent se baser exclusivement sur les documents sources fournis (notamment GIEC AR6, ODD de l'ONU, Taxonomie de l'UE, et IPBES). Ne génère pas d'informations hors de ces textes.

Mission :
Ton objectif est de construire une taxonomie hiérarchique stricte à 3 niveaux pour servir de base d'entraînement à nos algorithmes.
Niveau 1 : Le "Broad Subject" (Soit CLIMAT, soit BIODIVERSITÉ).
Niveau 2 : Le "Macro-Secteur" économique ou environnemental.
Niveau 3 : La "Sous-catégorie" litigieuse spécifique.

Pour l'ensemble des sources, je veux que tu extraies et structures l'information exactement selon le format suivant. Ne fais pas de phrases d'introduction, sors directement l'architecture :

[BROAD SUBJECT : CLIMAT ou BIODIVERSITÉ]

[MACRO-SECTEUR] (ex: Systèmes Énergétiques, Usage des Sols, Transports)

Sous-catégorie 1 : [Nom, ex: Extraction minière pour batteries]

Tags Multi-Label : [Assigner ici soit le Broad Subject principal, soit obligatoirement la balise "CLIMAT & BIODIVERSITÉ (TRADE-OFF)" si le sujet implique un compromis ou impacte les deux sphères simultanément].

Définition Conceptuelle : [Une phrase stricte définissant les limites de cette sous-catégorie selon les textes. Si c'est un Trade-Off, explique brièvement pourquoi le sujet chevauche le climat et la biodiversité].

Acteurs & Intérêts Clés : [Liste des types d'entreprises, lobbys, ONG ou secteurs économiques dont les intérêts sont liés à ce contentieux].

Mots-clés NLP (Cibles pour classification) :

Institutionnel/Technique : [3 à 5 mots en anglais utilisés dans la loi/science]

Médiatique/Grand public : [3 à 5 mots en anglais couramment utilisés dans la presse]

Controverse/Polarisation : [3 à 5 mots en anglais liés aux doutes, critiques, scandales ou oppositions]

Sous-catégorie 2 : [Nom]

[... même structure ...]

Contraintes de génération :

Si un Macro-Secteur ou une Sous-catégorie appartient aux deux Broad Subjects de manière égale (ex: Agriculture intensive), place-le physiquement sous le Broad Subject où son impact principal est le plus documenté dans les textes, mais utilise impérativement le Tag "CLIMAT & BIODIVERSITÉ (TRADE-OFF)" pour que notre base de données NLP puisse le traiter en Multi-Label.

Les "Mots-clés NLP" doivent IMPÉRATIVEMENT être en Anglais (car nos modèles analyseront des datasets anglophones), même si ta définition est en français.

Élimine les mots trop génériques (comme "nature", "green", "bad") qui fausseraient un algorithme de classification. Choisis des entités nommées ou des bigrammes spécifiques (ex: "carbon capture", "deep sea mining").
```

## Output

Our output for the keywords is in a CSV file: contentious subjects classification.csv
To make sure the keywords are exported in the format adapted to the classification algorithm, we used Gemini 3.1 Pro to create a code to transorm the CSV file in the required list for the algorithm.
This code is csv_to_dictionnary.py

```python
import pandas as pd

# 1. Configuring and Reading the CSV File
file_path = "/workspaces/5452_DS-AI_Project_TeamI/PreTreatment/contentious subjects classification.csv"
# Using `quotechar` and `on_bad_lines` to prevent reading errors
df = pd.read_csv(file_path, sep=';', quotechar='"', on_bad_lines='skip')

# Intermediate dictionary for storing cleaned data
topics = {}

# 2. Data Extraction and Cleaning
for index, row in df.iterrows():
    sub_cat = row['(Sub-Category)']
    nlp_words_raw = row['Key NLP words targets']
    
    # Rows where the subcategory or keywords are empty are ignored
    if pd.isna(sub_cat) or pd.isna(nlp_words_raw):
        continue
        
    sub_cat = str(sub_cat).strip()
    words_str = str(nlp_words_raw).strip()
    
    # Removing the enclosing quotation marks generated by the CSV format
    while words_str.startswith('"') and words_str.endswith('"'):
        words_str = words_str[1:-1]
        
    # Separation by comma and individual cleaning of each word
    # We remove spaces and quotation marks (single or double) around each word
    words_list = [w.strip().strip('"\'') for w in words_str.split(',')]
    words_list = [w for w in words_list if w] # Exclusion of potential empty elements
    
    topics[sub_cat] = words_list

# 3. Formatting and writing the result (exactly as requested)
# We will write the result to a text/python file so that we can easily retrieve it
output_file = "/workspaces/5452_DS-AI_Project_TeamI/PreTreatment/topics_extracted.py"

with open(output_file, "w", encoding="utf-8") as f:
    f.write("topics = {\n")
    
    items = list(topics.items())
    for i, (sub_cat, words) in enumerate(items):
        f.write(f'    "{sub_cat}": [\n')
        
        # Formatting with line breaks (e.g., 4 words per line for readability)
        chunk_size = 4
        for j in range(0, len(words), chunk_size):
            chunk = words[j:j+chunk_size]
            # Add quotation marks around each word
            chunk_formatted = ", ".join([f'"{w}"' for w in chunk])
            
            # If this is not the last line of words for this sub-category, add a comma
            if j + chunk_size < len(words):
                f.write(f'        {chunk_formatted},\n')
            else:
                f.write(f'        {chunk_formatted}\n')
                
        # If this is not the last sub-category in the dictionary, close with a comma and skip a line
        if i < len(items) - 1:
            f.write('    ],\n\n')
        else:
            f.write('    ]\n')
            
    f.write("}\n")

print(f"Extraction completed successfully! The result has been saved in '{output_file}'.")
```
