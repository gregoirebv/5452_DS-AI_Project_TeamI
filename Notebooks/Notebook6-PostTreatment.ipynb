### Post-treatment of the dataset to include the relevant tags/information in each article in PostTreatment_dataset_CorEx.csv

```python
# Loading both CSVs and checking for columns
import pandas as pd

df1 = pd.read_csv("PostTreatment_dataset_CorEx.csv", sep=";")
df2 = pd.read_csv("../PreTreatment/contentious subjects classification.csv", sep=";")

df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

print("1.csv columns:", df1.columns.tolist())
print("2.csv columns:", df2.columns.tolist())
```

```python
# Defining the key column for CSVs comparison

key_col = "(Sub-Category)"

# Excluding "Sub-Category" and "Key NLP words targets" for columns copy
cols_from_2 = [
    "Broad subject",
    "Category / Contentious sector",
    "Category Conceptual Definition",
    "Sub-Category Conceptual Definition",
    "Key interests / key lobbies"
]

df2_reduced = df2[[key_col] + cols_from_2]

print("2.csv columns:", df2_reduced.columns.tolist())
```

```python
# Merging back into PostTreatment_dataset_CorEx.csv

df1_enriched = df1.merge(
    df2_reduced,
    on=key_col,
    how="left"
)

# 7. Save the enriched version back into 1.csv
df1_enriched.to_csv("PostTreatment_dataset_CorEx.csv", sep=";", index=False)
```
